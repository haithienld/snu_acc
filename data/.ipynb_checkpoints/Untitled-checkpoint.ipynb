```python
new = df['video'].str.split(".", n = 4, expand = True)
```

```python
df["userID"] = new[1]
df["startTime"] = new[2]
df["endTime"] = new[3]
```

```python
old_cols = df.columns.values 
old_cols
```

```python
new_cols= ['video', 'startTime', 'endTime', 'userID','View','label_action']
df = df.reindex(columns=new_cols)
```

```python
g = df.groupby(['View', df.label_action])
```

```python
import os

for (view, label_action), group in g:
    if not os.path.exists(str(view)):
        os.makedirs(str(view))
    group.iloc[:,:].to_csv(f'{view}/{label_action}.csv', index=False)
```
