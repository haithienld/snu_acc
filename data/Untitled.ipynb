```python
import pandas as pd
import os
```

```python
# Read train data
df = pd.read_csv('annotations/val.csv',sep=' ',names=['fileName',"labelClass"], header=None)
```

```python
df.loc[df['fileName'].str.contains('Dashboard') == True, 'cView'] = "1"
df.loc[df['fileName'].str.contains('Right') == True, 'cView'] = "2"
df.loc[df['fileName'].str.contains('Rear') == True, 'cView'] = "3"
# df.loc[df['video'].str.contains('Dashboard') = True, 'view']
# if(df["video"].str.contains('Right')):
#     df["view"] = 1
# if(df["video"].str.contains('Rear')):
#     df["view"] = 0
```

```python
df
```

```python
new = df['fileName'].str.split(".", n = 4, expand = True)
```

```python
df["userID"] = new[1]
df["startTime"] = new[2]
df["endTime"] = new[3]
```

```python
old_cols = df.columns.values 
old_cols
```

```python
new_cols= ['fileName', 'startTime', 'endTime', 'userID','cView','labelClass']
df = df.reindex(columns=new_cols)
```

```python
df
```

```python
g = df.groupby(['cView', df.labelClass])
```

```python
import os

for (cView, labelClass), group in g:
    if not os.path.exists("camView"+str(cView)):
        os.makedirs("camView"+str(cView))
    group.iloc[:,:].to_csv(f'{"camView"+cView}/{labelClass}.csv', index=False)
```
